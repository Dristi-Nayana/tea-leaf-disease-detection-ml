```python
import os
import cv2
import numpy as np
from skimage import feature, color

# Function to extract texture features (Local Binary Patterns)
def extract_texture_features(gray_image):
    lbp = feature.local_binary_pattern(gray_image, 8, 1, method='uniform')
    hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 10), range=(0, 10))
    texture_features = hist / np.sum(hist)
    return texture_features

# Function to extract color features (RGB Mean)
def extract_color_features(rgb_image):
    r_mean = np.mean(rgb_image[:, :, 0])
    g_mean = np.mean(rgb_image[:, :, 1])
    b_mean = np.mean(rgb_image[:, :, 2])
    return [r_mean, g_mean, b_mean]

# Function to extract size features (Leaf Area)
def extract_size_features(gray_image):
    _, thresh = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    leaf_area = cv2.contourArea(contours[0])
    return leaf_area

# Function to extract shape features (Perimeter, Circularity, Aspect Ratio)
def extract_shape_features(contour):
    perimeter = cv2.arcLength(contour, True)
    (x, y), (w, h), angle = cv2.minAreaRect(contour)
    leaf_area = cv2.contourArea(contour)
    circularity = (4 * np.pi * leaf_area) / (perimeter ** 2)
    aspect_ratio = w / h
    return perimeter, circularity, aspect_ratio

# Path to the folder containing leaf images
folder_path = '/content/drive/MyDrive/tea_sickness_dataset/white spot'

# Iterate over each image in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Load image
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Extract features
        texture_features = extract_texture_features(gray)
        color_features = extract_color_features(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        size_features = extract_size_features(gray)
        contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if len(contours) > 0:
            shape_features = extract_shape_features(contours[0])
        else:
            shape_features = (0, 0, 0)  # Default values if no contour is found

        # Printing or saving the extracted features
        print("File:", filename)
        print("Texture Features (Local Binary Patterns):", texture_features)
        print("Color Features (RGB Mean):", color_features)
        print("Size Features (Leaf Area):", size_features)
        print("Shape Features:")
        print("- Perimeter:", shape_features[0])
        print("- Circularity:", shape_features[1])
        print("- Aspect Ratio:", shape_features[2])
        print("\n")
```
